```python
from splinter import Browser
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
import pandas as pd
```

```python
def init_browser():
    return Browser()
```

```python
browser=Browser()
```

```python
url = 'https://www.realtor.com/realestateandhomes-search/Houston_TX'
browser.visit(url)
```

```python
prices = []
streets = []
zipcodes = []
cities = []
```

```python
for x in range(1,6):#tqdm_notebook(range(1, 6)):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    browser.visit(f'https://www.realtor.com/realestateandhomes-search/Houston_TX/pg-{x}')
    
    listing = soup.find_all('li', class_='component_property-card js-component_property-card js-quick-view')# js-component_property-card js-quick-view')
#     print(listing)  
    for l in listing:
        price = l.find('span', class_='data-price')
        zipcode = l.find('span', class_='listing-postal')
        street = l.find('span', class_='listing-street-address')
        city = l.find('span', class_='listing-city')
        print('-------------')
        print(price.text)
        print(street.text)
        print(zipcode.text)
        print(city.text)
        prices.append(price.text)
        streets.append(street.text)
        zipcodes.append(zipcode.text)
        cities.append(city.text)
```

```python
house_data = pd.DataFrame({'Price':prices,'Street':streets,'Zip Code':zipcodes,'City':cities})
```
