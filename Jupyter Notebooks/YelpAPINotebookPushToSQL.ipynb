```python
# Get Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Get Pickled Restaurant List
restaurants = pd.read_pickle('business_result_df.pkl')
```

```python
# Restaurants List Appears to Contain Restaurants Outside Houston due to Zip Code Searches Returning International Results

# Filter Alias to Only Houston Containing Aliases
restaurants = restaurants.loc[restaurants['alias'].str.contains('houston')].reset_index(drop=True)
restaurants
```

```python
# Connect to DB
rds_connection_string = 'wen:12345@localhost:5432/yelp_db'
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Get Table Names
table_names = engine.table_names()
table_names
```

```python
# Check if restaurants is in tables
if 'restaurants' in table_names:
    # Get Existing Restaurants
    restaurants_existing = pd.read_sql_query('select * from restaurants', con=engine)
    # Only Add Restaurants Not in Existing
    add_restaurants = restaurants.loc[~restaurants['alias'].isin(restaurants_existing['alias'])].copy()
    # Add Restaurants
    add_restaurants.to_sql(name='restaurants', con=engine, if_exists='append', index=False)
# If Not in Tables, Add to Tables    
else:
    restaurants.to_sql(name='restaurants', con=engine, if_exists='append', index=False)
```
