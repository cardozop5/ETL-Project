```python
from Keys import yelp_id, yelp_api_key
import requests
import pprint as pp
import pandas as pd
```

```python
# Get List of Zip Codes
path = '../Raw Data/ZipCodes.csv'
zip_codes = pd.read_csv(path)
zip_codes.head()
```

```python
# Filter For Zip Codes in Texas
zip_codes_tx = zip_codes.loc[zip_codes['state_abv']=='TX'].copy().reset_index(drop=True)
zip_codes_tx.head()

# Filter Only For Houston
zip_codes_hou_tx = zip_codes_tx.loc[zip_codes_tx['city']=='Houston'].copy().reset_index(drop=True)
zip_codes_hou_tx.head()
```

```python
# Get Number of Zip Codes
print(zip_codes_hou_tx.shape[0])

# Get Zips Into a List
zip_codes_search = list(zip_codes_hou_tx['zip_code'])
zip_codes_search
```

```python
headers = {'Authorization': 'Bearer ' + yelp_api_key}
url = 'https://api.yelp.com/v3/businesses/search'
business_results = []


zips_found = 0
# Loop Through Zip Codes
for zip_code in zip_codes_search:
    try:
        offset = 0
        params = {'location': zip_code,
                 'term': 'Restaurants',
                 'limit': 50,
                 'offset': offset}

        response = requests.get(url=url, params=params, headers = headers)

        #If Request Successful- Write to List of Dictionaries
        if response.status_code == 200:
            results = response.json()
            # If Results List is Empty
            if not business_results: 
                business_results = results['businesses']
            else:
                business_results.extend(results['businesses'])
            # Get Number of Results
            results_count = results['total']

        #If Request Results Are More Than Limit- Offset Another 50
        if results_count > 50:

            #While offset + 50 < result_count 
            while offset +50 < results_count:
                offset += 50
                params = {'location': zip_code,
                         'term': 'Restaurants',
                         'limit': 50,
                         'offset': offset}
                response = requests.get(url=url, params=params, headers = headers)

                #If Request Successful- Write to List of Dictionaries
                if response.status_code == 200:
                    results = response.json()
                    # If Results List is Empty
                    if not business_results: 
                        pp.pprint(business_results)
                        business_results = results['businesses']
                    else:
                        business_results.extend(results['businesses'])

        zips_found += 1
        print(f'{zips_found} zips found')
    except:
        print('Zip Code Not Found... Move On')
```

```python
# Convert results to a DataFrame
business_result_df = pd.DataFrame(business_results)
business_result_df
```

```python
# Keep Subset (alias, categories, name, price, rating, review_count)
business_result_df = business_result_df[['alias', 'name', 'price', 'rating', 'review_count', 'location']]

# Drop Duplicates (alias)
business_result_df = business_result_df.drop_duplicates(subset = ['alias']).reset_index(drop=True)
business_result_df
```

```python
# Extract Zip Code
business_result_location_list = business_result_df['location'].tolist()
business_result_location_df = pd.DataFrame(business_result_location_list)
business_result_location_df = business_result_location_df[['zip_code']]
```

```python
# Get Back Into 1 DF
business_result_df = pd.concat([business_result_df, business_result_location_df], axis=1)
del business_result_df['location']
```

```python
# Drop NAs In zip_code 
business_result_df = business_result_df.dropna(subset=['zip_code']).reset_index(drop=True)
```

```python
# Save to Pickle (as Backup)
business_result_df.to_pickle('business_result_df.pkl')
```
